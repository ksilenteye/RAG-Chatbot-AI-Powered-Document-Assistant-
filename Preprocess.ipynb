```python
import fitz  # PyMuPDF
from langchain.text_splitter import RecursiveCharacterTextSplitter
from pathlib import Path
```

```python
def extract_pdf_text(pdf_path):
    """
    Extracts and returns all text from a PDF file as a single string.
    Args:
        pdf_path (str): Path to the PDF file.
    Returns:
        str: Extracted text.
    """
    try:
        doc = fitz.open(pdf_path)
        full_text = ''
        for page in doc:
            full_text += page.get_text()
        return full_text.strip().replace('\n', ' ')
    except Exception as e:
        print(f'Error reading PDF: {e}')
        return ''
```

```python
def chunk_text(text, chunk_size=300, overlap=50):
    """
    Splits text into overlapping chunks.
    Args:
        text (str): The text to split.
        chunk_size (int): Size of each chunk.
        overlap (int): Overlap between chunks.
    Returns:
        list: List of text chunks.
    """
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=chunk_size,
        chunk_overlap=overlap,
        length_function=len
    )
    return splitter.split_text(text)
```

```python
def save_chunks(chunks, save_path='chunks/chunks.txt'):
    """
    Saves text chunks to a file, separated by --- lines.
    Args:
        chunks (list): List of text chunks.
        save_path (str): Output file path.
    """
    try:
        Path(save_path).parent.mkdir(exist_ok=True, parents=True)
        with open(save_path, 'w', encoding='utf-8') as f:
            for chunk in chunks:
                f.write(chunk + '\n---\n')
        print(f'Successfully saved {len(chunks)} chunks to {save_path}')
    except Exception as e:
        print(f'Error saving chunks: {e}')
```

```python
raw_text = extract_pdf_text(PDF_PATH)
if raw_text:
    chunks = chunk_text(raw_text, chunk_size=CHUNK_SIZE, overlap=CHUNK_OVERLAP)
    save_chunks(chunks, OUTPUT_PATH)
    print(f'{len(chunks)} chunks saved.')
else:
    print('No text extracted from PDF.')
```
